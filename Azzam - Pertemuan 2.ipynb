```python
def jumlah_data(x):
    n = len(x)
    jumlah = 0
    for i in range(n):
        jumlah = jumlah + x[i]
    return jumlah
```

```python
datax = [5, 4, 2, 3]
jumlah_data(datax)
```

```python
def cek_bilangan(bil):
    if bil < 0:
        print("Bilangan Negatif")
    elif bil > 0:
        print("Bilangan Positif")
    else:
        print("Nol")
```
